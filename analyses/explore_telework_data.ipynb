## basic line chart using Altair

```python
def make_line(df:pd.DataFrame, x_col: str, y_col: str, color:str = None, title = None) -> alt.Chart:
    line_chart = alt.Chart(df).mark_line(point=True).encode(
    alt.X(x_col),
    alt.Y(y_col),
    alt.Color(color) if color else alt.value("black"),
    tooltip=[x_col, y_col, color]
    )
    if title:
        return line_chart.properties(title=title)
    else:
        return line_chart              
```

## state_hybrid_workforce_trend

```python
melt_hybrid_tr = hybrid_tr.melt(
    id_vars="Report Month",
    var_name="employee_category",
    value_name="employee_count"
)

display(
    melt_hybrid_tr.describe(include="all"),
    melt_hybrid_tr["employee_category"].value_counts()
)
```

```python
make_line(melt_hybrid_tr, 
          x_col="Report Month", 
          y_col="employee_count", 
          color="employee_category",
          title="Hybrid Workforce Trends"
         )
```
